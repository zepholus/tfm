```python
import pandas as pd
import plotly.express as px
import glob
```

```python
root = './observacions/'
path_observacions_filtrat = root+'observacions_filtrat/'
estacions_path = root+"estacions_cabal.csv"
```

```python
estacions = pd.read_csv(estacions_path, index_col=5)
estacions.head()
```

```python
for observacions_path in glob.glob(path_observacions_filtrat+"*.csv")[:5]:
    nom = observacions_path.replace(path_observacions_filtrat, '').replace('.csv', '')
    estacio = estacions.loc[nom]['Estació']
    observacions = pd.read_csv(observacions_path)
    fig = px.line(observacions, x='Date', y="Flow", title=estacio)
    fig.show()
```
