```python
from config.config import SWAT_PREDICTIONS_DIR
import pandas as pd
import os
```

```python
#if file swat_predictions.csv exists, delete it
if os.path.exists(SWAT_PREDICTIONS_DIR / 'swat_predictions.csv'):
    os.remove(SWAT_PREDICTIONS_DIR / 'swat_predictions.csv')
```

```python
#read all csv files in SWAT_PREDICTIONS_DIR

swat_predictions = []

for f in SWAT_PREDICTIONS_DIR.glob('*.csv'):
    print(f)
    df = pd.read_csv(f)

    #drop date column
    df.drop('date', axis=1, inplace=True)

    swat_predictions.append(df)
```

```python
#add columns based on index
swat_predictions_merged = pd.concat(swat_predictions, axis=1)
```

```python
swat_predictions_merged
```

```python
#add date columnm, begginning in 1/1/2001
swat_predictions_merged['datetime'] = pd.date_range(start='1/1/2001', periods=len(swat_predictions_merged), freq='D')
swat_predictions_merged.set_index('datetime', inplace=True)
```

```python
#sort columns by name
swat_predictions_merged = swat_predictions_merged.sort_index(axis=1)
swat_predictions_merged
```

```python
#delate rows from 2022
swat_predictions_merged = swat_predictions_merged[:'2021-12-31']
swat_predictions_merged
```

```python
swat_predictions_merged.to_csv(SWAT_PREDICTIONS_DIR / 'swat_predictions.csv')
```
