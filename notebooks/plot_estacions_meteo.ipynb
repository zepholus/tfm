```python
from config.config import DATA_MODEL_DIR, ESTACIONS_DIR, SWAT_PREDICTIONS_DIR, LSTM_DATA_DIR
import pandas as pd
import hvplot.pandas # noqa: adds hvplot method to pandas objects
from holoviews.element.tiles import EsriImagery
import geopandas as gpd
import holoviews as hv
```

```python
precipitation_stations_path = DATA_MODEL_DIR / 'swat files 2022' / 'pcp.txt'
temperature_stations_path = DATA_MODEL_DIR / 'swat files 2022' / 'tmp.txt'
```

```python
df = pd.read_csv(precipitation_stations_path)
df2 = pd.read_csv(temperature_stations_path)
df
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.LONG, df.LAT)
)

gdf
```

```python
a = gdf.hvplot.points(x='LONG', y='LAT', geo=True, c='red', tiles='EsriImagery').opts(width = 700, height = 700)
```

```python
b = df2.hvplot.points(x='LONG', y='LAT', geo=True, tiles='EsriImagery', c = 'red').opts(width = 700, height = 700)
```

```python
map = a + b
```
